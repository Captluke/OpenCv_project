```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
from tensorflow.keras.optimizers import RMSprop

import pandas as np
import numpy as np

import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
```

```python
path = 'C:/Users/Lukas/Desktop/computer-vision/basedata/training/mask/1.jpg'
training_path = 'C:/Users/Lukas/Desktop/computer-vision/basedata/training/'
validation_path = 'C:/Users/Lukas/Desktop/computer-vision/basedata/validation/'

img = image.load_img(path)
plt.imshow(img)
```

```python
cv2.imread(path).shape # 194=pixel height, 259=pixel width, 3 rgb color
```

```python
train = ImageDataGenerator(rescale=1/255)
validation = ImageDataGenerator(rescale=1/255)
```

```python
train_dataset = train.flow_from_directory(
                                        training_path,
                                        target_size=(200,200), ##resize the image to given size as 200 x 200
                                        batch_size= 3,
                                        class_mode='binary')

validation_dataset = validation.flow_from_directory(
                                        validation_path,
                                        target_size=(200,200), ##resize the image to given size as 200 x 200
                                        batch_size= 3,
                                        class_mode='binary'
                                         )
```

```python
model = tf.keras.models.Sequential([tf.keras.layers.Conv2D(16,(3,3),activation = 'relu', input_shape=(200,200,3)), tf.keras.layers.MaxPool2D(2,2),
                                    #
                                    tf.keras.layers.Conv2D(32,(3,3),activation = 'relu'),
                                    tf.keras.layers.MaxPool2D(2,2),
                                    #
                                    tf.keras.layers.Conv2D(64,(3,3),activation = 'relu'), 
                                    tf.keras.layers.MaxPool2D(2,2),
                                    #
                                    tf.keras.layers.Flatten(),
                                    #
                                    tf.keras.layers.Dense(512, activation='relu'),
                                    #
                                    tf.keras.layers.Dense(1,activation='sigmoid')
                                    ]) # We have to define our model by using convulational network
```

```python
# This section we compile the model 
model.compile(loss='binary_crossentropy',
              optimizer = RMSprop(lr=0.001),
              metrics = ['accuracy'] )
```

```python
# This section we run to train the dataset (Train dataset and Validation Dataset only)
model_fit = model.fit(train_dataset,
                    steps_per_epoch=3,
                    epochs=40,
                    validation_data=validation_dataset)
```

```python
# This section is to test whether the train model works well against Testing dataset (Just for verification)
dir_path = 'C:/Users/Lukas/Desktop/computer-vision/basedata/testing'
for i in os.listdir(dir_path):
    #print(i)
    img = image.load_img(dir_path+'//'+i, target_size=(200,200,3))
    plt.imshow(img)
    plt.show()

    X = image.img_to_array(img)
    X = np.expand_dims(X, axis=0)
    images = np.vstack([X]) # Join all the images
    val = model.predict(images)
    # Refer to this {'mask': 0, 'no_mask': 1}
    if val == 0:
        print("with Mask")
    elif val == 1:
        print("No Mask")
```
